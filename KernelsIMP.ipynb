```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(-5.0,5.0,100)
y=np.sqrt(10**2-x**2)
x=np.hstack([x,-x])  #stack horizontally on existing axis
y=np.hstack([y,-y])
```

```python
y
```

```python
x1=np.linspace(-5.0,5.0,100)
y1=np.sqrt(5**2-x1**2)
x1=np.hstack([x1,-x1])
y1=np.hstack([y1,-y1])
```

```python
y1
```

```python
plt.scatter(x,y)
plt.scatter(x1,y1)
plt.show()
```

```python
np.vstack([y,x]).T #it will stack vertically on the existing vertical axis 
```

```python
import pandas as pd 
df1=pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2=pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df=pd.concat([df1,df2]) #by using the concat we have stacked both df1 and df2 
df
```

```python
df.head()
```

```python
df.tail()
```

```python
#independent and dependent features 
X=df.iloc[:,:2]
y=df.Y
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

implementing the polynomial kernel


```python
#requires x1 square , x2 square and x1*x2 in the df 
df['X1_Square']=df['X1']**2
df['X2_Square']=df['X2']**2
df['X1*X2']=(df['X1']*df['X2'])
```

```python
df.head()
```

```python
#independent and dependent feaature 
X=df[['X1','X2','X1_Square','X2_Square','X1*X2']]
y=df.Y
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
import plotly.express as px

fig = px.scatter_3d(df, x='X1', y='X2', z='X1*X2',
              color='Y')
fig.show()
```

```python
fig = px.scatter_3d(df, x='X1_Square', y='X1_Square', z='X1*X2',
              color='Y')
fig.show()
```

now we can do the classification as the data are not overlappin witheachother 

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="poly")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="rbf")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel="sigmoid")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
